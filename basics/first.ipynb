```python
import matplotlib.pyplot as plt

a=[]
b=[]

for x in range(-50,50,1):
    y=x**2+2*x+2
    a.append(x)
    b.append(y)

fig= plt.figure()
axes=fig.add_subplot(111)
axes.plot(a,b)
plt.show()
```
